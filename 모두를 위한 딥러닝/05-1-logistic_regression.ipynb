```python
import tensorflow as tf
```

```python
x_data = [[1, 2],
          [2, 3],
          [3, 1],
          [4, 3],
          [5, 3],
          [6, 2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]
```

```python
# sigmoid를 활성함수로 하고, binary_crossentropy를 loss함수로 사용하는 모델
tf.model1 = tf.keras.Sequential()
tf.model1.add(tf.keras.layers.Dense(units=1, input_dim=2))
tf.model1.add(tf.keras.layers.Activation('sigmoid'))
tf.model1.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model1.summary()
```

```python
# sigmoid를 활성함수로 하고, mse를 loss함수로 사용하는 모델
tf.model2 = tf.keras.Sequential()
tf.model2.add(tf.keras.layers.Dense(units=1, input_dim=2))
tf.model2.add(tf.keras.layers.Activation('sigmoid'))
tf.model2.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model2.summary()
```

```python
history1 = tf.model1.fit(x_data, y_data, epochs=500)
```

```python
history2 = tf.model2.fit(x_data, y_data, epochs=500)
```

```python
print("Model1_Accuracy: ", history1.history['accuracy'][-1])
```
